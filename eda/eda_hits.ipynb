маппинг категории и экшенов по ивентам

```python
import pandas as pd

# === 1. Важные действия пользователя ===
important_actions = [
    'view_card', 'view_new_card', 'quiz_show', 'quiz_start',
    'go_to_car_card', 'sub_landing', 'sub_view_cars_click', 'photos_all',
    'start_chat', 'user_message', 'chat established', 'client initiate chat',
    'click_on_start_chatting', 'click_on_request_call', 'form_request_call_sent',
    'sub_submit_success', 'callback requested', 'mobile call',
    'sub_callback_submit_click', 'sub_car_request_submit_click',
    'click_buy_auto', 'click_sell_car', 'click_credit', 'click_pos_credit',
    'click_on_credit_btn', 'click_setelem_credit', 'setelem_credit_form_button',
    'tinkoff_credit_form_button', 'sravni_credit_form_button',
    'from_credit_landing_pos', 'survey_buy_after_sell_start',
    'survey_buy_after_sell_when', 'click_car_buyback',
    'click_on_subscription', 'auto_subscription_click',
    'open_offer', 'sub_offer_click'
]

# === 2. Сбор action-фичей по этим действиям ===
filtered_hits = df_hits[df_hits['event_action'].isin(important_actions)]
action_counts = filtered_hits.groupby(['session_id', 'event_action']).size().unstack(fill_value=0)

# Бинарные признаки
for action in important_actions:
    if action not in action_counts.columns:
        action_counts[action] = 0
    action_counts[f'has_{action}'] = (action_counts[action] > 0).astype(int)

# Целевой признак
action_counts['target'] = (
    (action_counts.get('sub_submit_success', 0) > 0) |
    (action_counts.get('form_request_call_sent', 0) > 0)
).astype(int)

action_features = action_counts.reset_index()

# === 3. Поисковые фильтры (search_form_*) ===
search_actions = df_hits[df_hits['event_action'].str.startswith('search_form_')]
search_features = (
    search_actions.groupby('session_id')['event_action']
    .agg([
        ('num_search_events', 'count'),
        ('num_search_types', 'nunique')
    ])
)
search_features['has_used_filters'] = (search_features['num_search_events'] > 0).astype(int)

# === 4. event_label фичи ===
label_features = (
    df_hits
    .dropna(subset=['event_label'])
    .groupby('session_id')['event_label']
    .agg([
        ('num_labels', 'count'),
        ('num_unique_labels', 'nunique')
    ])
)

# === 5. event_category фичи ===
category_counts = (
    df_hits
    .dropna(subset=['event_category'])
    .groupby(['session_id', 'event_category'])
    .size()
    .unstack(fill_value=0)
)
for col in category_counts.columns:
    category_counts[f'has_cat_{col}'] = (category_counts[col] > 0).astype(int)

# === 6. Объединяем всё ===
df_model = (
    action_features
    .merge(search_features, on='session_id', how='left')
    .merge(label_features, on='session_id', how='left')
    .merge(category_counts, on='session_id', how='left')
)
df_model.fillna(0, inplace=True)

df_model.to_csv("df_model_full.csv", index=False)

```

Проверяем заранее соотношение таргета к городам, чтобы не удалить лишнего:

```python
# Считаем общее количество таргетов
total_targets = df_merged['target'].sum()

# Считаем количество таргетных сессий по городам
target_by_city = (
    df_merged[df_merged['target'] == 1]
    .groupby('geo_city')['target']
    .count()
    .rename('target_count')
)

# Считаем общее количество сессий по городам
total_by_city = (
    df_merged
    .groupby('geo_city')['session_id']
    .count()
    .rename('total_count')
)

# Объединяем
city_stats = pd.concat([target_by_city, total_by_city], axis=1).fillna(0)

# Вычисляем доли
city_stats['percent_of_targets'] = (city_stats['target_count'] / total_targets * 100).round(2)
city_stats['percent_of_all_sessions'] = (city_stats['total_count'] / df_merged.shape[0] * 100).round(2)

# Сортируем по вкладу в таргеты
city_stats_sorted = city_stats.sort_values(by='percent_of_targets', ascending=False)

# Показываем топ-20
city_stats_sorted.head(20)
```

Москва: даёт 48% всех таргетов при 43% всех сессий — вклад в таргеты выше среднего.

Санкт-Петербург: даёт 13.1% таргетов при 16.1% сессий — чуть ниже среднего по эффективности.

Остальные города: вклад в таргеты уже <3%, а большинство даже <1%.
geo_city как raw-признак слишком раздут.

Но топовые города с отклонением по таргету — информативны.

```python
# Выделим топ-10 городов по количеству таргетов
top_10_cities = city_stats_sorted.head(10).index.tolist()

# Функция группировки
def group_city(city):
    if city in top_10_cities:
        return city
    elif city == '(not set)':
        return 'unknown'
    else:
        return 'other'

# Добавим geo_city в df_merged через merge, если его нет
if 'geo_city' not in df_merged.columns:
    df_merged = df_merged.merge(df_sessions[['session_id', 'geo_city']], on='session_id', how='left')

# Создание признака
df_merged['geo_city_grouped'] = df_merged['geo_city'].apply(group_city)

df_merged.head(20)
```

```python
import pandas as pd

# === 1. Загрузка CSV-файлов ===
df_model = pd.read_csv('df_sessions_cleaned.csv')
df_merged = pd.read_csv('df_hits_cleaned.csv')

# === 2. Кодировка visit_number ===
def encode_visit_number(n):
    try:
        n = int(n)
        if n == 1:
            return '1'
        elif n == 2:
            return '2'
        elif n == 3:
            return '3'
        elif n == 4:
            return '4'
        else:
            return '5+'
    except:
        return 'unknown'

df_merged['visit_number_encoded'] = df_merged['visit_number'].apply(encode_visit_number)

# === 3. Объединение таблиц по session_id ===
df_final = df_model.merge(df_merged, on='session_id', how='left', suffixes=('', '_merged'))

# (по желанию можно посмотреть список всех колонок)
print(df_final.columns.tolist())

# === 4. Проверка результата ===
print(df_final.shape)
df_final.head()
```
