```python
import pandas as pd

import pandas as pd

# 1. –ó–∞–≥—Ä—É–∂–∞–µ–º —Ñ–∞–π–ª—ã
df_model = pd.read_csv('df_cleaned_for_model.csv')     # —Å–µ—Å—Å–∏–æ–Ω–Ω—ã–µ –∞–≥—Ä–µ–≥–∞—Ç—ã
df_hits = pd.read_csv('ga_hits.csv')                   # —Å—ã—Ä—ã–µ —Ö–∏—Ç—ã —Å event_action

# 2. –û–ø—Ä–µ–¥–µ–ª—è–µ–º —Å–ø–∏—Å–æ–∫ —Ü–µ–ª–µ–≤—ã—Ö —Å–æ–±—ã—Ç–∏–π
target_actions = ['showed_number_ads', 'quiz_show']

# 3. –û—Ç–º–µ—á–∞–µ–º –≤ df_hits, –µ—Å—Ç—å –ª–∏ —Ü–µ–ª–µ–≤–æ–µ –¥–µ–π—Å—Ç–≤–∏–µ
df_hits['custom_target'] = df_hits['event_action'].isin(target_actions).astype(int)

# 4. –ì—Ä—É–ø–ø–∏—Ä—É–µ–º –ø–æ session_id: 1, –µ—Å–ª–∏ –≤ —Å–µ—Å—Å–∏–∏ –±—ã–ª–æ —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ —Ü–µ–ª–µ–≤–æ–µ —Å–æ–±—ã—Ç–∏–µ
session_targets = df_hits.groupby('session_id')['custom_target'].max().reset_index()

# 5. –ü—Ä–∏—Å–æ–µ–¥–∏–Ω—è–µ–º –∫ df_model
df_model = df_model.merge(session_targets, on='session_id', how='left')

# 6. –ü—Ä–æ–ø—É—Å–∫–∏ (—Ç.–µ. —Å–µ—Å—Å–∏–∏ –±–µ–∑ —Ü–µ–ª–µ–≤—ã—Ö —Å–æ–±—ã—Ç–∏–π) –∑–∞–º–µ–Ω–∏–º –Ω–∞ 0
df_model['custom_target'] = df_model['custom_target'].fillna(0).astype(int)

# 7. –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
print(df_model['custom_target'].value_counts(normalize=True))

```

–í—ã–±–µ—Ä–µ–º —Ç–æ–ª—å–∫–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ —Å–µ—Å—Å–∏–∏ (custom_target == 1)

–ü–æ–ª—É—á–∏–º —Å–æ–±—ã—Ç–∏—è (event_action) –ø–æ session_id –≤ –ø–æ—Ä—è–¥–∫–µ –≤—Ä–µ–º–µ–Ω–∏ (hit_number)

–°–≥—Ä—É–ø–ø–∏—Ä—É–µ–º –∏ —Å–æ–±–µ—Ä—ë–º –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏

–ü–æ—Å—á–∏—Ç–∞–µ–º —á–∞—Å—Ç—ã–µ –ø–∞—Ç—Ç–µ—Ä–Ω—ã



```python
# 1. –í—ã–±–µ—Ä–µ–º —Å–µ—Å—Å–∏–∏ —Å —Ü–µ–ª–µ–≤—ã–º –¥–µ–π—Å—Ç–≤–∏–µ–º
target_sessions = df_hits[df_hits['custom_target'] == 1]['session_id'].unique()

# 2. –û—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ –Ω—É–∂–Ω—ã–µ —Å—Ç—Ä–æ–∫–∏
df_seq = df_hits[df_hits['session_id'].isin(target_sessions)]

# 3. –£–ø–æ—Ä—è–¥–æ—á–∏–º –ø–æ session_id –∏ hit_number (—á—Ç–æ–±—ã —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –ø–æ—Ä—è–¥–æ–∫ –¥–µ–π—Å—Ç–≤–∏–π)
df_seq = df_seq.sort_values(['session_id', 'hit_number'])

# 4. –°–≥—Ä—É–ø–ø–∏—Ä—É–µ–º –ø–æ session_id –∏ —Å–æ–±–µ—Ä–µ–º –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Å–æ–±—ã—Ç–∏–π
session_sequences = df_seq.groupby('session_id')['event_action'].apply(list)

# 5. –ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —á–∞—Å—Ç—ã–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏
from collections import Counter

# –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Å–ø–∏—Å–∫–∏ —Å–æ–±—ã—Ç–∏–π –≤ —Å—Ç—Ä–æ–∫–∏ (–Ω–∞–ø—Ä–∏–º–µ—Ä, "view_card > sub_landing > quiz_show")
patterns = session_sequences.apply(lambda x: ' > '.join(x))

# —Å—á–∏—Ç–∞–µ–º —á–∞—Å—Ç–æ—Ç—É
pattern_counts = Counter(patterns)

# 6. –í—ã–≤–µ–¥–µ–º —Ç–æ–ø-10 —Å–∞–º—ã—Ö —á–∞—Å—Ç—ã—Ö –ø—É—Ç–µ–π
for path, count in pattern_counts.most_common(10):
    print(f"{count:5d} | {path}")
```

–£–¥–∞–ª–µ–Ω–∏–µ –ø–æ–≤—Ç–æ—Ä—è—é—â–∏—Ö—Å—è –¥–µ–π—Å—Ç–≤–∏–π (—Ü–∏–∫–ª–æ–≤)

```python
from collections import Counter

def remove_consecutive_duplicates(seq):
    result = []
    prev = None
    for action in seq:
        if action != prev:
            result.append(action)
        prev = action
    return result

# –ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏–º, session_sequences ‚Äî —ç—Ç–æ —Å–ø–∏—Å–æ–∫ —Å–ø–∏—Å–∫–æ–≤ –¥–µ–π—Å—Ç–≤–∏–π
cleaned_sequences = [remove_consecutive_duplicates(seq) for seq in session_sequences]

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –æ–±—Ä–∞—Ç–Ω–æ –≤ —Å—Ç—Ä–æ–∫—É –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ —á–∞—Å—Ç–æ—Ç
sequence_strings = [' > '.join(seq) for seq in cleaned_sequences]
sequence_counts = Counter(sequence_strings)

# –¢–æ–ø-10 —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–µ–π
for seq, count in sequence_counts.most_common(10):
    print(f"{count:5d} | {seq}")
```

2. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å Sankey-–¥–∏–∞–≥—Ä–∞–º–º—É
–î–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ –∫–ª—é—á–µ–≤—ã—Ö –ø–µ—Ä–µ—Ö–æ–¥–æ–≤ –º–µ–∂–¥—É –¥–µ–π—Å—Ç–≤–∏—è–º–∏. –ü—Ä–∏–º–µ—Ä –¥–ª—è plotly:

```python
import plotly.graph_objects as go

# –ü–æ–¥–≥–æ—Ç–æ–≤–∏–º –ø–∞—Ä—ã –ø–µ—Ä–µ—Ö–æ–¥–æ–≤
edges = []
for seq in cleaned_sequences:
    for i in range(len(seq) - 1):
        edges.append((seq[i], seq[i+1]))

# –°—á–∏—Ç–∞–µ–º —á–∞—Å—Ç–æ—Ç—É –ø–µ—Ä–µ—Ö–æ–¥–æ–≤
edge_counts = Counter(edges)
top_edges = edge_counts.most_common(30)  # –≤–æ–∑—å–º–µ–º —Ç–æ–ª—å–∫–æ 30 —Å–∞–º—ã—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ —Ñ–æ—Ä–º–∞—Ç –¥–ª—è Sankey
labels = list(set([item for pair in edge_counts for item in pair]))
label_to_index = {label: idx for idx, label in enumerate(labels)}

source = [label_to_index[src] for (src, tgt), _ in top_edges]
target = [label_to_index[tgt] for (src, tgt), _ in top_edges]
value = [count for _, count in top_edges]

fig = go.Figure(data=[go.Sankey(
    node=dict(label=labels),
    link=dict(source=source, target=target, value=value)
)])

fig.update_layout(title_text="Sankey: –ü–µ—Ä–µ—Ö–æ–¥—ã –º–µ–∂–¥—É –¥–µ–π—Å—Ç–≤–∏—è–º–∏", font_size=10)
fig.show()
```
